# FIFA World Cup

```python
import pandas as pd
```

Abrir el csv con datos del mundial

```python
wc = pd.read_csv('worldcup.csv')
```

```python
wc
```

```python
wc.shape
```

Resultados solo de Brasil 2014

```python
brasil = wc.loc[wc["Year"] == 2014]

brasil
```

```python
brasil_goles = brasil[["Home Team Name", "Away Team Name", "Home Team Goals", "Away Team Goals"]]

brasil_goles

brasil_goles.describe()
```

Imprimir la información de todas las finales

```python
finals = wc.loc[wc["Stage"] == "Final"]

# finals

finals[["Home Team Name", "Home Team Goals", "Away Team Name", "Away Team Goals", "Year"]]
```

Partidos con más de 10 goles

```python
muchos_goles = wc.loc[wc["Home Team Goals"] + wc["Away Team Goals"] >= 10]

muchos_goles
```

Victorias de argentina de local 🇦🇷

```python
argentina = wc.loc[(wc["Home Team Name"] == "Argentina") & (wc["Home Team Goals"] > wc["Away Team Goals"])]

argentina
```

```python
argentina.shape
```

Finales de Holanda 🇳🇱

```python
netherlands = wc.loc[(wc["Stage"] == "Final") & ((wc["Home Team Name"] == "Netherlands") | (wc["Away Team Name"] == "Netherlands"))]

netherlands
```

Es posible usar la función query para hacer más "natural" la búsqueda

```python
maracanasos_locales = wc.query("Stadium == 'Estadio do Maracana' and `Home Team Goals` > `Away Team Goals`")

maracanasos_locales
```

Se pueden utilizar ciclos para recorrer la información

```python
for i in range(len(wc)):
    if (wc.loc[i]["Stage"] == "Final"):
        print(wc.loc[i]["Home Team Name"] + " (" + str(wc.loc[i]["Home Team Goals"]) + ") vs " + wc.loc[i]["Away Team Name"] + " (" + str(wc.loc[i]["Away Team Goals"]) + ")")
```
